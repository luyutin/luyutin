```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
from sklearn.metrics import accuracy_score, confusion_matrix

# 1.將資料集讀入
df = pd.read_csv('titanic/train.csv')
df.head()
```

```python
# 2.將sex, embarked欄位做編碼
df['Sex'] = pd.Categorical(df['Sex']).codes
df['Embarked'] = pd.Categorical(df['Embarked']).codes
df.head()
```

```python
#3.移除含有缺失值之row
df = df.dropna()
```

```python
# 4.使用"Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"欄位來定義特徵向量X
# 5.使用"Survived"來定義y
selected_colunms = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
x = df[selected_colunms]
y = df['Survived']
```

```python
#6.將資料集 8:2分成訓練資料及測試資料(假設random_state=1)
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#7.將訓練資料做標準化
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
```

```python
#8.建立機器學習模型並訓練
GNB_model = GaussianNB()
GNB_model.fit(X_train,y_train)
```

```python
#9.將測試資料做標準化並用訓練好的模型做預測
X_test_standardized = scaler.transform(X_test)
y_pred = GNB_model.predict(X_test_standardized)
```

```python
#10.驗證模型結果
num_correct_samples = accuracy_score(y_test, y_pred, normalize = False)
accuracy = accuracy_score(y_test, y_pred)
con_matrix = confusion_matrix(y_test, y_pred)

print('number of correct sample: {}'.format(num_correct_samples))
print('accuracy: {}'.format(accuracy))
print('confusion matrix: {}'.format(con_matrix))
```
